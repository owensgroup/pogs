```python
def random_walk(n, start=0, p_inc=.2):
    return start + np.cumsum(np.random.uniform(size=n) < p_inc)
```

```python
starts = np.random.choice(range(4), 10)
probs = [.1, .3, .5]
walks = np.dstack([[random_walk(15, s, p) for s in starts] for p in probs])
```
